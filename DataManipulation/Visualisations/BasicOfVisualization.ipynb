```python
#import section
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('heart.csv')
```

***BAR PLOT***

```python
chest_pain = df.groupby(by='cp').median().reset_index()
```

```python
chest_pain.head()
```

```python
#pandas inbuild library
chest_pain.plot.bar(x='cp',y='age');
```

```python
chest_pain.plot.bar(x='cp');
```

```python
#matplotlib version
fig, ax = plt.subplots()
ax.bar(chest_pain.cp,chest_pain.age,label='age',color=['green','blue','pink','yellow'],edgecolor='k')
ax.set_xlabel('CP')
ax.set_ylabel('Age')
ax.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
#Seaborn version of born plot
sns.set_style('dark')
sns.barplot(data=df,x='cp',y='age',errcolor='white', label='age');
plt.legend();
```

**Scatter Plots***

```python
#pandas inbuild libraries
df.plot.scatter('age','trestbps',c='black');
```

```python
#Matplotlib version
fig, ax = plt.subplots()
ax.scatter(data=df,x='age',y='trestbps',marker='*',s=40,c='age',edgecolor='k',alpha=0.5,label='age')
ax.set_xlabel('Age')
ax.set_ylabel('trestbps')
ax.legend(bbox_to_anchor = (1,1))
plt.show()
```

```python
#Seaborn version of scatter plot
sns.set_style('white')
sns.scatterplot(x='age',y='trestbps',data=df,hue='age',edgecolor='k');
```

**LINE PLOT**

```python
ages = df.groupby('age').median().reset_index()
ages.head()
```

```python
ages.plot.line('age',['chol','cp','trestbps']);
```

```python
#Pandas version of line
fig, ax = plt.subplots()
ax.plot(ages['age'],ages['chol'],ls='--')
ax.set_xlabel("age",fontsize=20)
ax.set_ylabel('Cholestrol',fontsize=20);
```

```python
#seaborn version of line
sns.lineplot('age','chol',data=ages);
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
fig, axes = plt.subplots(ncols=2,)
df.plot.scatter(x= 'age',y='chol',ax=axes[0])
df.plot.scatter(x= 'age',y='trestbps',ax=axes[1])
fig.tight_layout()
```

```python
#Access and save plot
fig2 = axes[1].get_figure()
fig.savefig('Output.png',transperent=True,bbox_inches= 'tight')
```

```python
with plt.style.context('default'):
    fig, axes = plt.subplots(ncols=3,sharey=True,gridspec_kw={"width_ratios":[2,2,2],"wspace":0})
    y  = 'age'
    xs = ['chol','trestbps','thalach']
    
    for x, ax in zip(xs,axes):
        ax.scatter(x= df[x],y=df[y])
        ax.set_xlabel(x)
        
    axes[0].set_ylabel(y)
    
#save the file
fig.savefig('output1.png',dpi=100,bbox_inches='tight')
fig.savefig('output1.pdf',dpi=100,bbox_inches='tight')
```
