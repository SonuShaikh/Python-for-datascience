```python
#import section
import pandas as pd
import numpy as np
import pickle

filename = 'heart.csv'
```

```python
#Best way to load database
df = pd.read_csv(filename)
df.head()
```

*Load text method*

```python
data= np.loadtxt(fname=filename,delimiter=',',skiprows=1)
print(data)
```

```python
data= np.genfromtxt(filename,delimiter=',',dtype=None,names=True,encoding='utf-8-sig')
data
```

```python
#Read file manually (Default option in pythong)
```

```python
def load_file(filename):
    with open(filename,encoding='utf-8-sig') as f:
        data, cols = [],[]
        for i, line in enumerate(f.read().splitlines()):
            if i == 0:
                cols += line.split(',')
            else:
                data.append([float(x) for x in line.split(',')])
            df = pd.DataFrame(data,columns=cols)
        return df

load_file(filename).head()
```

*** CREATE DATAFRAME AND SAVING THEM ***

```python
#DataFrame By Array
data = np.random.random(size=(5,3))
print('NumPy Array:')
print(data)

print('\n Pandas DataFrame')
df = pd.DataFrame(data=data,columns=['A','B','C'])
df
```

```python
#DataFrame By Dictionary
df = pd.DataFrame(data={"Age":[10,20,30],"Name":['Sam','Alex',"John"],"Class":[12,34,78]})
df
```

```python
#Dataframe by dtype
dtype = [("A",np.int),("B",(np.str,20))]
data = np.array([(1,'Sam'),(2,'Alex'),(3,'John')], dtype=dtype)
df = pd.DataFrame(data,)
df
```

**Saving and Serializing DataFarme**

```python
data = np.random.random(size=(1000,5))
df = pd.DataFrame(data, columns=['A','B','C','D','E'])
df.head()
```

```python
#save to CSV file
df.to_csv('save-numpy-to-dataframe.csv',index=False)
```

```python
#save to pickle file
df.to_pickle('save-picket-format.pkl')
```

```python
#Save in HDF (Hirarchical Data Format)
df.to_hdf('save-hirachcal-data-format.hdf',key='data',format='table')
```

```python
#Save in Feather format
df.to_feather('save-feather-format.fth')
```

In terms of file size, HDF5 is the largest for this example. Everything else is approximately equal. For small data sizes, often csv is the easiest as its human readable.
