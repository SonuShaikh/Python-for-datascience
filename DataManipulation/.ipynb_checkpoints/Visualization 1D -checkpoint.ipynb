**Histogram** <br/>
Most common was to show distribution

```python
#import section
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('heart.csv')
```

```python
#pandas version
df.age.plot.hist(bins=20.);
```

```python
#matplot version
plt.hist(df.age,bins=20);
```

```python
fig, ax = plt.subplots()
ax.hist(df.trestbps,bins=20,histtype='step',label='trestbps',edgecolor='r');
ax.hist(df.thalach,bins=20,histtype='stepfilled',label='thalach',alpha=0.3,edgecolor='g');
ax.legend();
```

**BOXPLOT** n\
Shows mean, Quratile, IRQ and outliner

```python
#pandas box version 
df[['trestbps','thalach']].plot.box();
```

```python
df[['trestbps','thalach']].describe()
```

```python
#matplotlibe version of 
plt.boxplot(df[['trestbps','thalach']].to_numpy());
plt.boxplot()
```

```python
#seaborn version of boxplot
```

```python
sns.boxplot(data=df[['trestbps','thalach']]);
```

```python
#seaborn version of
sns.boxplot(data=df,x='cp',y='trestbps');
```

```python
df.groupby('cp').boxplot(column='trestbps',sharex=True,layout=(1,4),grid=False);
plt.tight_layout()
```

### Violen Plot <br/>
Like box plots... but better. Not a staple of pandas at the moment(though it has been requested)

```python
fig, ax = plt.subplots()
ax.violinplot(df[['trestbps','thalach']].to_numpy());
ax.set_xlabel('Data',fontsize=20);
ax.set_ylabel('Numbers',fontsize=20);
```

```python
fig, ax = plt.subplots()
ax.violinplot(df[['trestbps','thalach']].to_numpy(),bw_method=0.2);
ax.set_xlabel('Data',fontsize=20);
ax.set_ylabel('Numbers',fontsize=20);
```

```python
#sea born version
sns.violinplot(data=df[['trestbps','thalach']],inner='quartile',bw=0.2);
```

**BEE SWARM PLOT** <Br/>
If you want to go fancy, these can be fun for presentation. but let concise than other plot. 

```python
sns.swarmplot(data=df[['trestbps','thalach']],size=3.5,color='g');
```

```python
sns.violinplot(data=df[['trestbps','thalach']],inner=None);
sns.swarmplot(data=df[['trestbps','thalach']],size=3.5,color='g',alpha=0.5);
```
