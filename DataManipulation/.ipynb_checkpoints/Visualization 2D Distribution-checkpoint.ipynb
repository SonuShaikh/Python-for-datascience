```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('meteorite-landings.csv')
```

```python
df.info()
```

```python
#Removing NA values for dataframes
df = df.dropna(subset=['reclat','reclong'])
df = df[df.reclong < 300]
```

**2D HISTOGRAM**

```python
plt.hist2d(df.reclong,df.reclat,bins=200,vmax=4);
plt.colorbar();
```

**2D HEX PLOT**

```python
plt.hexbin(df.reclong,df.reclat,bins=200,vmax=4,lw=0.0)
plt.colorbar();
```

```python
#pandas version of hexbin
df.plot.hexbin(x='reclong',y='reclat',vmax=2, gridsize=100,linewidth=None);
```

****CONTOUR***
\n A contour line of a function of two variables is a curve along which the function has a constant value, so that the curve joins points of equal value. It is a plane section of the three-dimensional graph of the function f parallel to the-plane.

```python
#Lets create some data
spacing = np.linspace(0,10,200)
x,y = np.meshgrid(spacing,spacing)
z = (np.sin(x)+np.cos(y) + 2 * np.arcsinh(x*y))**2
plt.contour(x,y,z,levels=20);
plt.colorbar();
```

```python
c =plt.contour(x,y,z,levels=20);
plt.clabel(c, inline=True,fmt="%0.1f",colors='black')
plt.colorbar();
```

```python
plt.contourf(x,y,z,levels=20);
plt.colorbar();
```

```python
plt.contourf(x,y,z,levels=20);
c = plt.contour(x,y,z,levels=20,colors='black');
plt.clabel(c, inline=True,fmt="%0.1f")
plt.colorbar();
```

***KDE Kernal Dencity Estination**

```python
#this is the called rejection sampling. a way to brute force sample any surface
n = 50000
xs, ys = np.random.uniform(0, 10, n), np.random.uniform(0, 10,n)
zs = (np.sin(xs) + np.cos(ys) + 2 * np.arcsinh(xs * ys))**2
zs /= zs.max()
passed = np.random.uniform(0,1,n) < zs
xs,ys = xs[passed],ys[passed]
plt.scatter(xs,ys, s=1, alpha=0.2);
```

```python
sns.kdeplot(xs,ys,bw=3.0,shade=True,shade_lowest=True);
sns.kdeplot(xs,ys,bw=3.0,shade_lowest=True);
```

```python
#JointPlot
sns.jointplot(data=df,x='reclong',y='reclat');
```

```python
sns.jointplot(data=df,x='reclong',y='reclat',kind='hex',gridsize=100,vmax=3,linewidth=0,marginal_kws={'bins':100});
```

```python
sns.jointplot(x=xs,y=ys, kind='kde')
```

```python
sns.jointplot(xs,ys,kind='hex',gridsize=20,cmap='magma');
```

```python
sns.pairplot(data=df[['reclong','reclat','mass']])
```
