### Concat And append

You've seen append before. We have grades for 2 semesters, and we want them both together

```python
grades1['semester'] = '1'
grades2['semester'] = '2'
```

```python
display(grades1.head(2),grades2.head(2))
```

```python
grades = pd.concat([grades1,grades2])
grades
```

```python
# If we don't want to append pandas concat then we can user append function
grades1.copy().append(grades2)
```

```python
grades = grades.reset_index(drop=True)
grades
```

##### That's better. Note that it would have caused us issues in this example. Now lets actually merge these dataframes togegher. Let's put the student information in , merging on the student id
### Merging horizontally

```python
#merge student df with grades

students_grades = pd.merge(students,grades,left_on='id',right_on='student_id')
print('Shape',students_grades.shape)
students_grades.head()
```

```python
#For some reasons we want use the on colums clause
students2 = students.rename({'id':'student_id'},axis=1)
students2.head()
```

```python
students_fullname = students2.merge(contacts,on='student_id')
students_fullname.head()
```

```python
students_grades = pd.merge(students2,grades,on='student_id')
df = pd.merge(students_grades,contacts,on='student_id')
df.head()
```

```python
df = df.merge(teachers,on='course')
df.head()
```

```python
#Lest generlized the data for all failed students

df2 = df.loc[df.grade == 'F','student_id']
method1 = df2.value_counts().rename_axis('student_id').reset_index(name='Counts')
method1
```

```python
method2 = df.loc[df.grade == 'F',['student_id','grade']].groupby('student_id').count().reset_index()
method2.columns = ['student_id','counts']
method2.sort_values('counts',ascending=False)
```

```python
#Let's find student have failed in 3 or more exams

stud_id = method2[method2.counts >= 3].student_id
stud_id
```

```python
final = students_fullname[students_fullname.student_id.isin(stud_id)]
final
```

### Join
what about pd.join? its a specific version for merge data that works solely more on index

```python
display(students.head(),contacts.head())
```

```python
students2 = students.set_index('id')
contacts2 = contacts.set_index('student_id')

students2.join(contacts2).head(5)
```

Quick tip: <br/>
We can use join and merge to combine 2 dataframes. The join method works best when we are joining dataframes on their indexes (though you can specify another column to join on for the left dataframe). The merge method is more versatile and allows us to specify columns besides the index to join on for both dataframes.

The other difference is you can join multple dataframes at once, provided they don't have overlapping columns.
