### Merging Functions
There are some functions for merging we haven't looked at yet, I've made up some tempreture data from us to play with.

* merge_order()
###### pandas.merge_ordered(left, right, on=None, left_on=None, right_on=None, left_by=None, right_by=None, fill_method=None, suffixes=('_x', '_y'), how='outer')
Perform merge with optional filling/interpolation.  
Designed for ordered data like time series data. Optionally perform group-wise merge. <br/><br/>

* merge_asof()
###### pandas.merge_asof(left, right, on=None, left_on=None, right_on=None, left_index=False, right_index=False, by=None, left_by=None, right_by=None, suffixes=('_x', '_y'), tolerance=None, allow_exact_matches=True, direction='backward')
Perform an asof merge.
This is similar to a left-join except that we match on nearest key rather than equal keys. Both DataFrames must be sorted by the key.

```python
import pandas as pd
import matplotlib.pyplot as plt

df_usa = pd.read_csv('usa.csv',parse_dates=['date'],dayfirst=True)
df_aus = pd.read_csv('aus.csv',parse_dates=['date'],dayfirst=True)
df_bris = pd.read_csv('brisbane.csv',parse_dates=['date'],dayfirst=True)

display(df_aus.head(),df_usa.head(),df_bris.head())
```

```python
plt.plot(df_aus.date,df_aus.temperature,label='aus');
plt.plot(df_usa.date,df_usa.temperature,label='usa');
plt.plot(df_bris.date,df_bris.temperature,label='bris');
plt.legend();
```

### Ordered merging
Designed for merging ordered data, like time series data. So what if we wanted to combine our two sources into one.

```python
pd.merge(df_aus,df_usa,on='date',how='outer')
```

```python
pd.merge_ordered(df_aus,df_usa,on='date',suffixes=['_aus','_usa']).head(10)
```

```python
#filling Nan with fill forward method.
pd.merge_ordered(df_aus,df_usa,on='date',suffixes=['_aus','_usa'],fill_method='ffill')
```

Tada! A very basic forward-filled interpolate merged and ordered dataframe.

### Merging with partial/closest matches
Lets say we want the huge amout of data in the Brisbane temperatures, but we only want the closest temperature to a given date in either the australia or USA tempereture.

```python
df_both = pd.merge_ordered(df_aus,df_usa,on='date',suffixes=['_aus','_usa'],fill_method='ffill')
df_all  = pd.merge_asof(df_both,df_bris,on='date').rename(columns={'temperature':'temperature_bris'})
df_all.plot('date',['temperature_aus','temperature_usa','temperature_bris'],);
```


This has allowed us to pick out the dates corresponding to other measurements, which in this case correspond to the middle of summer and winter. Notice the fill value on the right(green line). we cold remove that if we wanted.<br/>

tolerance: int or Timedelta, optional, default None <br/>
Select asof tolerance within this range; must be compatible with the merge index. <br/?

directionbackward (default), forward, or nearest <br/>
Whether to search for prior, subsequent, or closest matches.

```python
df_tmp  = pd.merge_asof(df_both,df_bris,on='date',tolerance=pd.Timedelta('14 days'),direction='nearest')
df_tmp.plot('date',df_tmp.columns[1:],legend=False);
```

For merge_asof yo merge on the ordered columns, like time or date. But if you want to still match other columns, there is left_by and right_by you can use.
Super useful utilities when the need arises for them. Doing it manually would be possible but painful and slow.
