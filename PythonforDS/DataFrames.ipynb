```python
#Data set for visualisations
import pandas as pd
data = pd.read_csv('C:\\Users\\shaik\\OneDrive\\Desktop\\Python Programming\P4-Demographic-Data.csv');
data.columns = (['CountryName', 'CountryCode', 'Birthrate', 'Internetusers','IncomeGroup'])
data.head()
```

```python
vis1 = sns.distplot(data.Internetusers,bins=10)
```

```python
vis2 = sns.distplot(data['Birthrate'], bins=5)
```

```python
#boxplot
vis3 = sns.boxplot(data=data,x='IncomeGroup',y='Birthrate')
```

```python
vis4 = sns.boxplot('IncomeGroup','Birthrate',data=data) 
```

```python
vis6 = sns.lmplot(data=data,x='Internetusers',y='Birthrate',fit_reg=False)
```

```python
#Color
vis7 = sns.lmplot(data=data,x='Internetusers',y='Birthrate',fit_reg=False, hue='IncomeGroup',size=6)
```
