```python
import requests
import bs4
```

```python
'http://books.toscrape.com/catalogue/category/books_1/index.html'
```

```python
'http://books.toscrape.com/catalogue/category/books_1/page-2.html'
```

```python
base_url = 'http://books.toscrape.com/catalogue/category/books_1/page-{}.html'
```

```python
base_url.format('2')
```

```python
base_url.format('3')
```

```python
base_url.format('4')
```

```python
res = requests.get(base_url.format('1'))
```

```python
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
len(soup.select('.product_pod'))
```

```python
first_page = soup.select('.product_pod')[0]
```

```python
#Grabbing title
first_page.select('a')[1]['title']
```

```python
#Checking the rating
'star-rating Three' in str(first_page)
```

```python
first_page.select('.star-rating.Three')
```

```python
[] != first_page.select('.star-rating.Two')
```

```python
# We can check if something is in two stars (string call in rating or example.select('.rating'))
#first_page.select('a')[1]['title'] to grab book title
```

```python
two_r_book_title = []
for n in range(1,51):
    scrap_url = base_url.format(n)
    res = requests.get(scrap_url)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    
    #Grabbing all books from pages
    books = soup.select('.product_pod')
    
    #Iterating through each book to check rating
    for book in books:
        if len(book.select('.star-rating.Two')) != 0:
            book_title = book.select('a')[1]['title']
            two_r_book_title.append(book_title)
```

```python
for title in two_r_book_title:
    print(title)
```
