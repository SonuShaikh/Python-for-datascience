```python
#Create functiont to plot graph

def plotGraph(num):
    for indx in range(num):
        plt.plot(Salary[indx],c='Blue',marker='o',label=Players[indx])
    plt.xticks(list(range(0,10)),Seasons,rotation='vertical')
    plt.legend(bbox_to_anchor=(1,1))
    plt.show()
```

```python
plotGraph(10)
```

```python
#Plot graph for games playe by each players
def plotGameGraph(num):
    for indx in range(num):
        plt.plot(Games[indx],c='Red',marker='s',label=Players[indx])
    plt.xticks(list(range(0,10)),Seasons,rotation='vertical')
    plt.legend(bbox_to_anchor=(1,1))
    plt.show
```

```python
plotGameGraph(3)
```

```python
def plotMplayed(players):
    for names in players:
        plt.plot(MinutesPlayed[Pdict[names]],c='red',ls='--',marker='s',label=names)
    plt.xticks(list(range(0,10)),Seasons,rotation='vertical')
    plt.legend(bbox_to_anchor=(1,1))
    plt.show()
```

```python
plotMplayed(['KobeBryant', 'JoeJohnson', 'LeBronJames', 'ChrisBosh', 'ChrisPaul', 'KevinDurant', 'DerrickRose', 'DwayneWade'])
```

```python
#Advance Function desing
def plotGraph(Players):
    col = {"KobeBryant":'Red',"JoeJohnson":'Green',"LeBronJames":'Blue',"CarmeloAnthony":'Cyan',"DwightHoward":'Red',"ChrisBosh":'Black',"ChrisPaul":'Cyan',"KevinDurant":'Yellow',"DerrickRose":'Magenta',"DwayneWade":'Yellow'}
    Mar = {"KobeBryant":'o',"JoeJohnson":'s',"LeBronJames":'>',"CarmeloAnthony":'<',"DwightHoward":'^',"ChrisBosh":'v',"ChrisPaul":'s',"KevinDurant":'p',"DerrickRose":'P',"DwayneWade":'*'}
    

    for names in Players:
        plt.plot(Salary[Pdict[names]],c=col[names],marker=Mar[names],label=names)
        plt.xticks(list(range(0,9)),Seasons,rotation='vertical')
        plt.legend(bbox_to_anchor=(1,1))

```
